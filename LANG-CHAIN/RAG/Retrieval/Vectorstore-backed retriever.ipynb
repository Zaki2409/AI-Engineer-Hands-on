## how to create a runnable retrieval object.

If you navigate to Langshans documentation, you'll find many retrievers suitable for different tasks.

We won't study these retrievers individually.

I'll leave it to you to find the one that works best for your project.

Rather, we'll create a runnable retriever object to plug it in as part of an expression language chain.



```python
import config

from langchain_openai.embeddings import OpenAIEmbeddings
from langchain_chroma import Chroma  # New way
from langchain_core.documents import Document
```

```python
embedding = OpenAIEmbeddings(model="text-embedding-ada-002",
                            openai_api_key = config.api_key)
vectorstore = Chroma( persist_directory= './vectore representrtion folder',
                                   embedding_function= embedding
                                   )
```

```python
len(vectorstore.get()['documents'])
```

```python
retriver = vectorstore.as_retriever(search_type ='mmr',
                                   search_kwargs={'k':3 , 'lambda':0.7})
```

```python
retriver #It's an instance of the vector store retriever class.
```

```python
question = "What software do data scientists use?"
```

```python
retrived_docs = retriver.invoke(question)
```

```python
retrived_docs
```

```python
for i in retrived_docs:
    print(f"Page Contennt: {i.page_content} \n---------\n Lecture title:{i.metadata['Lecture Title']}\n")
```

The invoke method returned the same chunks we got when applying the MMR search method in the last lesson.
