## In this lesson, we'll show how to add, update and delete documents within the vector store.

```python
import config

from langchain_openai.embeddings import OpenAIEmbeddings
from langchain_chroma import Chroma  # New way
from langchain_core.documents import Document
```

```python
embedding = OpenAIEmbeddings(model="text-embedding-ada-002",
                            openai_api_key = config.api_key)
```

```python
vectorstore_from_dir = Chroma( persist_directory= './vectore representrtion folder',
                                   embedding_function= embedding
                                   )
```

```python
vectorstore_from_dir.get() #As a result, we retrieve a long dictionary containing information for each embedding.
```

```python
vectorstore_from_dir.get(ids ="2b59d12b-2b0c-469a-b517-05da0c7698ae", include =["embeddings"])
```

####  how to add a document to the database.

```python
added_doc = Document(page_content='''Alright! So… Let’s discuss the not-so-obvious differences between the terms analysis and analytics. Due to the similarity of the words, some people believe they share the same meaning, and thus use them interchangeably. Technically, this isn’t correct. There is, in fact, a distinct difference between the two. And the reason for one often being used instead of the other
                                is the lack of a transparent understanding of both. So, let’s clear this up, shall we? First, we will start with analysis''',
                    metadata={'Course Title': 'INTRODUCTION', 'Lecture Title': 'Analysis vs Analytics'}
                        )
```

```python
vectorstore_from_dir.add_documents([added_doc])
```

```python
vectorstore_from_dir.get("c910ac07-92e3-42a4-a62a-d1f10dc49356")
```

#### All right, let's now update it with the document from the second lecture.

```python
updated_doc = Document (
page_content='''More importantly, it will be sufficient for your need to create quick and accurate analyses. However, if your theoretical preparation
is strong enough, you will find yourself restricted by software. Knowing a programming language such as R and Python, gives you the freedom to 
create specific, ad-hoc tools for each project you are working on''',
     metadata={'Course Title': 'INTRODUCTION', 'Lecture Title': 'Programming Languages & Software Employed in Data Science - All the Tools You Need'}
)
```

```python
vectorstore_from_dir.update_document(document_id= "c910ac07-92e3-42a4-a62a-d1f10dc49356",
                                    document= updated_doc)
```

```python
vectorstore_from_dir.get("c910ac07-92e3-42a4-a62a-d1f10dc49356")
```

#### deleting an object with a specified ID.

```python
vectorstore_from_dir.delete("c910ac07-92e3-42a4-a62a-d1f10dc49356")
```

```python
vectorstore_from_dir.get("c910ac07-92e3-42a4-a62a-d1f10dc49356")
```
