## Indexing: Document loading with Docx2txtLoader

```python
from langchain_community.document_loaders import Docx2txtLoader
```

```python
loader_docx= Docx2txtLoader("../Data/Introduction_to_Data_and_Data_Science.docx")
```

```python
pages_docx = loader_docx.load()
```

```python
pages_docx #We get a list with only one document object, this time storing the entire text from the file.
```

Scrolling down, we see that the stored metadata is only the path to the file without the page number.
Still, the page number would not be advantageous for our use case, since the text represents a transcript
of a video course.
A student asking a question wouldn't reference a specific page.
