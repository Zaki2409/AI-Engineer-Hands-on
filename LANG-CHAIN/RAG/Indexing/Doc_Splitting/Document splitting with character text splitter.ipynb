### Next comes the splitting procedure 

```python
from langchain_text_splitters.character import CharacterTextSplitter
```

```python
#By default, separator is set equal to two newline characters. A common way of separating titles from paragraphs.
#But we removed these characters from our text. So let's set the separator equal to an empty string and continue exploring other class parameters.

#Let's set the size of the chunks to 500 characters at maximum. By default, the value of the trunk overlap is set to 200 characters, so if you don't want any overlap
#between the splits, specify this explicitly. 
char_splitter = CharacterTextSplitter(separator ="", chunk_size = 500, chunk_overlap = 0)
```

```python
#One of the methods that this class implements is split documents It takes a list of document objects, such as. The pages list splits them accordingly and returns a new list of documents.
pages_char_split = char_splitter.split_documents(pages_docx)
```

### Let's now introduce an overlap between the chunks and see how this affects the results.


```python
char_splitter2 = CharacterTextSplitter(separator ="", chunk_size = 500, chunk_overlap = 50)
```

```python
pages_char_split2 = char_splitter2.split_documents(pages_docx)
```

Indeed, the final few characters of the first split were added as the beginning of the next one to
avoid abruptly ending chunks, sometimes even mid-word.
We can set the separator as a period indicating the end of a sentence.

```python
char_splitter3 = CharacterTextSplitter(separator =".", chunk_size = 500, chunk_overlap = 50)
pages_char_split3 = char_splitter3.split_documents(pages_docx)
pages_char_split3
```

```python
len(pages_char_split3)
```

```python
pages_char_split3[0].page_content
```

Retrieving the page content.

We see that this time the paragraph doesn't end abruptly, but at the end of the sentence instead.

You might also notice that the last period in the chunk is missing in each split, because separators

aren't included in the content.
