```python
# Upon inspecting the file, many new lines have appeared during the transcribing process.
pages_pdf = loader_pdf.load() #The result is a list of documents.
pages_pdf
```

```python
pages_num = len(pages_pdf)
pages_num
```

All right, let's now discuss the new lines introduced during the transcribing process.

One disadvantage of having so many redundant newline characters \n  is that they consume tokens.

As you know, token usage is monetized.

```python
#A parameter called page content stores the allocated portion of the text.
#The second available parameter metadata stores the metadata for a specific chunk of text.
#In our case, that's the document's path and the page number where the particular text is located.
```

```python
pages_pdf_cut = copy.deepcopy(pages_pdf) #  deep copy in this case to avoid overwriting the original file.
```

```python
#Okay, we now want to change the page content of the documents inside pages PDF cut by removing all newline characters.
#retrive the first contect of the page
first_content = pages_pdf_cut[0].page_content
```

```python
#We can apply the split method since the resulting object is a string.
list1 = first_content.split()
# this approach creates a list of all words in the sentence, omitting the newline characters.
```

```python
' '.join(list1) #This line of code joins all words in the list, with the character in the string serving as a separator.
```

```python
# We now must apply this procedure to all documents in the list using a for loop.

for i in pages_pdf_cut:
    i.page_content = ' '.join(i.page_content.split())
```

```python
pages_pdf_cut  # now no new line chartters
```

Each text file you work with requires special care and pre-processing.

In this example, the newline characters cluttered the text rather than contributed to the formatting,

so we're forced to remove them.

In the more frequent cases where the text is well formatted, the new lines can later be used to split

the text into meaningful chunks.
