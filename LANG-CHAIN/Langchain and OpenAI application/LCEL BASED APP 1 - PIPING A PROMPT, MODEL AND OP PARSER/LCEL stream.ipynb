We'll dedicate this lesson to the third and final method included in the expression language's standard

interface stream.



```python
import config
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
```

```python
chat_template = ChatPromptTemplate.from_messages(['human', "I've recently adopted a {pet} which is a {breed}. Could you suggest some training tips? \n"])
```

```python
chat = ChatOpenAI(
    model = "gpt-3.5-turbo",
    temperature = 0,
    seed = 365,
    max_tokens = 30,
    openai_api_key = config.api_key
)
```

```python
chain = chat_template | chat
```

```python
chain.invoke({'pet':'dog', 'breed': 'sheperd'})
```

```python
chain.batch([{'pet':'dog', 'breed': 'sheperd'},
            {'pet':'dragon', 'breed': 'night fury'}])
```

```python
chain.stream({'pet':'dog', 'breed': 'sheperd'})
```

### To quickly demonstrate what happens once we iterate through all generator items, let's decrease the
### maximum number of completion tokens to five.

```python
chat2 = ChatOpenAI(
    model = "gpt-3.5-turbo",
    temperature = 0,
    seed = 365,
    max_tokens = 5,
    openai_api_key = config.api_key
)
```

```python
chain2 = chat_template | chat
```

```python
#store the generator object in a variable called response.
#To access the elements in a generator one at a time, we set it as an argument to the built in Python
#function called next.
```

```python
response = chain2.stream({'pet':'dog', 'breed': 'sheperd'})
```

```python
next(response) #Let's execute the cell several times until we exhaust all five tokens.
```

```python
#Ultimately, we hit a stop iteration exception, signaling that the response generator has exhausted
#all its elements.
#It can no longer be used.
```
