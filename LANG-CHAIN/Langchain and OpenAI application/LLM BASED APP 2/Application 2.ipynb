##  we discuss and experiment with Lang Chan's chat messages.

```python
import config
from langchain_openai import ChatOpenAI


from langchain_core.messages import SystemMessage, HumanMessage, AIMessage
```

```python
chat = ChatOpenAI(
    model="gpt-3.5-turbo",
    temperature=0,
    seed=365,
    max_tokens =50,
    openai_api_key=config.api_key #LangChain's OpenAI integrations require the API key to be passed explicitly or via the environment variable.
)
```

## 4) Send Messages And Print The Result

LangChain chat models accept a list of messages (system, human, AI, tool messages). Here we provide a **system** instruction and a **human** prompt.

```python
message_system = SystemMessage(
    content=("You are Marv, a chatbot that reluctantly answers questions with sarcastic responses.")
)

message_human = HumanMessage(
    content=("I've recently adopted a dog. Can you suggest some dog names?")
)
```

```python
response= chat.invoke([message_system, message_human]) #define the response variable so that the list inside the invoke method includes the system message
print(response.content)
```

Last time we demonstrated the purpose of a system and a human message, storing chatbot instructions

and user requests, respectively.

In this lesson, we'll learn how to incorporate another type of chat message the AI message.

Recall that you've already encountered it as an assistant role message in OpenAI's API.

Remember that our response variables were also instances of Lang Chain's AI message class, so AI messages

are ones generated by the model.

They can also be used in few shot prompting.

Few shot prompting is a technique whereby we give the model a set of example responses to learn from,

so that it later generates responses that align better with our expectations.

And that's precisely the use case we'll demonstrate in this lesson.

First, remove the system message class from the list of imports.

Then the system message instance which instructs the bot to be sarcastic and lastly exclude this object

from the invoke methods argument.

This way we won't tell the model how to behave.

Instead, we'll teach it to give sarcastic responses through the few shot prompting technique.


message_human2_dog = HumanMessage(content = "I've recently adopted a dog. Can you suggest some dog names?" )
message_ai_dog = AIMessage(content = "Oh, how exciting! How about naming your dog Fido or Spot? Those are definitely unique and original choices.")
message_human2_cat = HumanMessage(content = "I've recently adopted a cat. Can you suggest some cat names?" )
message_ai_cat = AIMessage(content= ''' Congratulations on your new furry friend! Here are some cat name suggestions: Luna, Oliver, Bella, Simba, Cleo, Milo, Luna, Max, Chloe, and Leo.
''')
message_human2_fish = HumanMessage(content="I've recently adopted a fish. Can you suggest some fish names?")



```python
response2 = chat.invoke([message_human2_dog,message_ai_dog,message_human2_cat,message_ai_cat,message_human2_fish])
print(response2.content)
```

## Of course, providing the model with more examples will improve its responses even further.

## But remember that this comes at the cost of using up more prompt tokens.

These message objects proved very useful but have a significant drawback they are not reusable.

The developers at Lang Chain have therefore implemented prompt abstractions called chat prompt templates,

allowing for the reuse of system, human and AI messages.
