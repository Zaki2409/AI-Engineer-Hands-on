## OPEN AI's CHAT -- INVOKE() --- CHAT PROMPT TEMPLATE AND CHAT PROMTPT VLAUE CAN BEHANDLES BY INVOKE()

```python
import config

from langchain_openai import ChatOpenAI
from langchain_core.prompts.chat import (SystemMessagePromptTemplate,
                                         HumanMessagePromptTemplate,
                                        ChatPromptTemplate)
```

```python
chat = ChatOpenAI(
    model="gpt-3.5-turbo",
    temperature=0,
    seed=365,
    max_tokens =50,
    openai_api_key=config.api_key #LangChain's OpenAI integrations require the API key to be passed explicitly or via the environment variable.
)
```

```python
TEMPLATE_SYSTEM = '{description}'
TEMPLATE_HUMAN =''' 
    I've recently adopted a {pet}.
Could you suggest some {pet} names?    
    '''
```

```python
message_template_system = SystemMessagePromptTemplate.from_template(template = TEMPLATE_SYSTEM)
message_template_human = HumanMessagePromptTemplate.from_template(template = TEMPLATE_HUMAN)
```

```python
message_template_system
```

```python
message_template_human
```

```python
Chat_template = ChatPromptTemplate.from_messages([message_template_system, message_template_human])
```

```python
Chat_template
```

```python
Chat_value = Chat_template.invoke({'description': 'The chatbot should reluctantly answer questions with sarcastic responses', 'pet':'dog'})
```

```python
Chat_value
```

```python
#chat openai's invoke method accepts not only a single string or a list of chat messages,

response = chat.invoke(Chat_value)
print(response.content)
```

summary of our implementation and key observations that will

prove essential later in the course.

We first defined one system and one human string template.

Next.

Utilizing the from template method, we created an instance of the system and the human message prompt.

Template classes.

In the following step, a chat prompt template object combines these two instances.

We invoked it passing a dictionary as an argument.

The dictionary sets each input variable to its corresponding value.

The result was a chat prompt value object, which we passed as an argument to chat OpenAI's invoke method

to create a response in the form of an AI message.

So I hope you're starting to see the pattern here.

The output of one invoke method serves as the input to another.

Understanding this concept will be the stepping stone toward understanding how chains in Erlang chain

work.

### The Chaining Pattern
### Prompt template – you invoke it with variables, and it returns a “chat value object” (e.g., a list of BaseMessage objects, which are essentially message objects with roles and content).

### Chat model (e.g., ChatOpenAI) – you invoke it with that chat value object, and it returns a chat message (typically an AIMessage).
### That AIMessage can then be fed back into a prompt template (or directly into another model call) to build multi‑step chains.
