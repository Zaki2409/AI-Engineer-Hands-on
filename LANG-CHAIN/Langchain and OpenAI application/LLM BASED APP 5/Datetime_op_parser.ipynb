```python
import langchain
import langchain_core

print('langchain:', langchain.__version__)
print('langchain_core:', langchain_core.__version__)
```

```python
import config
from langchain_openai import ChatOpenAI
from langchain_core.messages import HumanMessage
from langchain_core.output_parsers import DatetimeOutputParser
```

```python
chat = ChatOpenAI(
    model = "gpt-3.5-turbo",
    temperature = 0,
    seed = 365,
    max_tokens = 30,
    openai_api_key = config.api_key
)
```

```python
message_human = HumanMessage(content=f''' When was the Danish poet Pete Heine born?''')
```

```python
response = chat.invoke([message_human])
print(response.content)
```

```python
date_output_parser = DatetimeOutputParser()
```

```python
responsed_parser =date_output_parser.invoke(response)
responsed_parser
```
