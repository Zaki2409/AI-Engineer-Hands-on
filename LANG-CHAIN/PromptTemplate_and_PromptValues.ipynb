## PROMPT TEMPLATES VS PROMPT VALUES

```python
from langchain_core.prompts import PromptTemplate
```

```python
# Some developers adhere to the convention of using all caps for string templates.
TEMPLATE = '''
System:
{description}

Human:
I've recently adopted a {pet}.
Could you suggest some {pet} names?
'''
```

```python
prompt_template = PromptTemplate.from_template(template =TEMPLATE)
```

```python
prompt_template
```

```python
prompt_value = prompt_template.invoke({'description': 'The chatbot should reluctantly answer questions with sarcastic responses', 'pet':'dog'} ) #this method expects a dictionary where we specify the values of the input variables.
```

```python
prompt_value
```

```python
print(prompt_value.text)
```
