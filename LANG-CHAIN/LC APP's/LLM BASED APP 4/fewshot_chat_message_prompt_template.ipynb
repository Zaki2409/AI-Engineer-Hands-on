# FEW -SHOT CHAT MESSAGE PROMPT TEMPLATE

### Let's now revisit the lesson where we created two human AI message pairs, which taught the model to
### answer questions sarcastically.
### We refer to this as few shot prompting, giving the model a couple of examples to learn from.
### This would be a zero shot prompting if we provided no examples to the model.
### I think we can all agree that this previous implementation of few shot prompting was somewhat clumsy.
### Our approach was to define a separate human eye messages pair for each example.
### This needlessly cluttered the code and is evidently hardly scalable.
### In this lesson, we'll utilize a class called Few Shot Chat Message prompt template and recreate the
### same example, but implemented much cleaner and more compactly.

```python
import config 
from langchain_openai import ChatOpenAI
from langchain_core.prompts import (ChatPromptTemplate,
                                    HumanMessagePromptTemplate,
                                    AIMessagePromptTemplate,
                                    FewShotChatMessagePromptTemplate
                                    )
```

```python
chat = ChatOpenAI(
    model = "gpt-3.5-turbo",
    temperature=0,
    seed = 365,
    max_tokens=50,
    openai_api_key = config.api_key
)
```

```python
TEMPLATE_H =''' 
I've recently adopted a {pet}.
Could you suggest some {pet} names?
'''
TEMPLATE_AI = ''' 
{response}
'''

message_template_human = HumanMessagePromptTemplate.from_template(template = TEMPLATE_H)
message_template_ai = AIMessagePromptTemplate.from_template(template = TEMPLATE_AI)
```

```python
#Combine both objects into a chat prompt template instance, which I'll call example template.
#Abiding by its name, this object will serve as the blueprint for our human AI example pairs.
example_template = ChatPromptTemplate.from_messages([message_template_human,message_template_ai])
```

```python
#What follows is creating a list of dictionaries, which I'll call examples, with each dictionary storing
#the values for a single example.
exmaple =  [{
    'pet':'dog',
     'response':'Oh, how exciting! How about naming your dog "Fido" or "Spot"? Those are definitely unique and original choices.'
},{
                'pet' : 'cat',
            'response' : 'Of course! Here are some cat name suggestions: Bubbles, Finley, Splash, Nemo, Coral, Sushi, Goldie, Neptune, Pearl, and Splash.' }
           ]
```

Now on the agenda is combining the chat prompt template object with the examples list in a single variable.


```python
few_shot_prompt = FewShotChatMessagePromptTemplate(examples = exmaple , example_prompt = example_template, input_variables=['pet'])
```

```python
chat_template = ChatPromptTemplate.from_messages([few_shot_prompt,message_template_human])
```

```python
chat_value = chat_template.invoke({'pet' :'rabbit'})
```

```python
chat_value
```

```python
for i in chat_value.messages:
    print(f'{i.type} : {i.content}\n')
```

The first four messages are part of the few shot examples, and the fifth comes from the human message

template.

```python
response= chat.invoke(chat_value)
```
