# Datetime Output Parser (LangChain 1.2.x)

## Notes On The LangChain 1.2.x Update

- `ChatOpenAI` is imported from `langchain_openai` (the package entry point) for a more stable import path.
- `seed` is passed explicitly to avoid warnings about values being routed through `model_kwargs`.
- The version check uses Python (instead of a shell command) so it runs consistently across Jupyter environments.


```python
import langchain
import langchain_core

print('langchain:', langchain.__version__)
print('langchain_core:', langchain_core.__version__)
```

```python
%load_ext dotenv
%dotenv
```

```python
from langchain_openai import ChatOpenAI
from langchain_core.messages import HumanMessage
from langchain_core.output_parsers import DatetimeOutputParser
```

```python
chat = ChatOpenAI(
    model='gpt-4o-mini',
    seed=365,
    temperature=0,
    max_tokens=100,
)
```

## Create A Message With Format Instructions

```python
parser = DatetimeOutputParser()

message_h = HumanMessage(
    content=f"""When was the Danish poet Piet Hein born?
{parser.get_format_instructions()}
"""
)
```

```python
print(message_h.content)
```

## Invoke The Model

```python
response = chat.invoke([message_h])
print(response.content)
```

## Parse The Model Output Into A Python `datetime`

```python
response_parsed = parser.invoke(response)
response_parsed
```
