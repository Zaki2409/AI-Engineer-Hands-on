```python
import config
from langchain_openai import ChatOpenAI
from langchain_core.messages import HumanMessage
from langchain_core.output_parsers import CommaSeparatedListOutputParser
```

```python
chat = ChatOpenAI(
    model = "gpt-3.5-turbo",
    temperature = 0,
    seed = 365,
    max_tokens = 30,
    openai_api_key = config.api_key
)
```

```python
message_human = HumanMessage(content=''' I've recently adopted a dog.

Could you suggest some dog names?
''')
```

 What we're aiming for is to have all names as elements of a list object.

```python
response = chat.invoke([message_human])
print(response.content)
```

```python
list_output_parser = CommaSeparatedListOutputParser()
```

```python
responsed_parser =list_output_parser.invoke(response)
responsed_parser
```

The response parsed a variable is indeed a list, but not the one we expected.

We wanted to have each name as a separate element.

Instead, we obtained the entire content as a single element.

#### So how can we arrive at the desired result?

#### Well, by instructing the chat model.



```python
print(list_output_parser.get_format_instructions())
```

Redefine the content as an F string and create an instance of the parser class within curly braces.
Then call the get format instructions method, run the cell and enter a new one to extract the content
of the human message.

```python
message_human2 = HumanMessage(content= f''' I've recently adopted a dog. Could you suggest some dog names?
{CommaSeparatedListOutputParser().get_format_instructions()}
''')
```

```python
print(message_human2.content)
```

```python
response2 = chat.invoke([message_human2])
print(response2.content)
```

```python
responsed_parser2 =list_output_parser.invoke(response2)
responsed_parser2
```
