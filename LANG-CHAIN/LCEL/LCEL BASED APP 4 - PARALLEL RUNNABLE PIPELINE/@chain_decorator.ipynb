We previously introduced Lang Chains Runnable Lambda class, which allowed upgrading a function to a

runnable object.

We demonstrated this using Python's anonymous functions, but you can wrap any function this way.

As we'll explain in a minute.

Additionally, we'll introduce lang chains decorators.



```python
from langchain_core.runnables import RunnableLambda
```

```python
def find_sum(x):
    return sum(x)
def find_sqr(x):
    return x**2
```

Now let's demonstrate how to create a chain using these two functions as building blocks.

```python
chain1 = RunnableLambda(find_sum) | RunnableLambda(find_sqr)
```

```python
chain1.invoke([14,54,343])
```

```python
from langchain_core.runnables import chain
```

```python
@chain   # Decorators are functions that modify the behavior of another function or a class, enhancing their functionality.
def runnable_sum(x):
    return sum(x)
@chain
def runnable_sqr(x):
    return x**2
#Here, the chain decorator wraps the runnable some function in a runnable lambda, just as we manually did above.
```

```python
type(runnable_sum), type(runnable_sqr)
```

```python
chain2 = runnable_sum | runnable_sqr
```

```python
chain2.invoke([14,54,343])
```

When using the chain decorator in your code, avoid naming your variables the same way.

Doing so introduces a name conflict in the code which could result in an error.
 
chain1 = RunnableLambda(find_sum) | RunnableLambda(find_sqr) # So what we're attempting to do in this cell is call the variable chain as if it were a Python function. But it's not. It's a runnable sequence object.
