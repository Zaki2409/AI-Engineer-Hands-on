## Now we'll continue expanding our toolbox and learn how to transform ordinary functions into runnable objects.
### For that, we'll use Python's Lambda functions.

The agenda for today's lesson is as follows.

First, we'll create two simple lambda functions, one that finds the sum of the elements inside a list,

and one that calculates the square of a number.

We'll then wrap them inside a runnable Lambda class, allowing us to call them using the invoke method.

We'll then construct a simple chain from those two objects and graph the result.



```python
find_sum = lambda x : sum(x)
find_sum([1,2,3])
```

```python
find_sqr= lambda x : x**2
find_sqr(3)
```

The functions we've just defined are great, but have one disadvantage.

We can't pipe them into a chain because they aren't runnable objects and don't implement the invoke

method.

To remedy this, we'll need to employ the main character of this lesson.

Runnable lambda.

```python
from langchain_core.runnables import RunnableLambda
```

```python
runnnable_sum = RunnableLambda(lambda x : sum(x))
```

```python
runnnable_sum.invoke([1,2,3])
```

```python
runnnable_sqr = RunnableLambda(lambda x : x**2)
```

```python
runnnable_sqr.invoke(34)
```

Now that we have these two runnable objects, let's pipe them together to construct a runnable sequence.

```python
chain =runnnable_sum | runnnable_sqr
```

```python
chain.invoke([1,2,3]) #The result is the square of the sum of these number
```

```python
chain.get_graph().print_ascii() #library that allowed for visualizing chain nodes and their sequence through an Ascii print.
```

I hope you can appreciate how powerful this runnable Lambda class is.

It allows us to convert any function into a runnable that we can then invoke and enter as a chain component.
