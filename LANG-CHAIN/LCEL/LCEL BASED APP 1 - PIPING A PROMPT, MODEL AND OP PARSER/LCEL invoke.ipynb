```python
import config
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import CommaSeparatedListOutputParser
```

```python
list_instruc = CommaSeparatedListOutputParser().get_format_instructions() #defined a variable called list instructions, where the respective formatting instructions are stored as a string.
list_instruc
```

What follows is instantiating a chat prompt template variable.
We've done this quite a few times already.
Now let's explore an alternative method for crafting the same template as previously discussed.


```python
chat_template = ChatPromptTemplate.from_messages(['human', "I've recently adopted a {pet}. Could you suggest three {pet} names? \n"])
```

```python
#The messages parameter, in turn, is a list storing a human message prompt template object.
#Notice that we didn't need to import and create an instance of this class.
#Conveniently, it was done for us.
chat_template 
```

```python
# Next, let's append the formatting instructions to the template.
chat_template1 = ChatPromptTemplate.from_messages([('human', "I've recently adopted a {pet}. Could you suggest three {pet} names? \n" + list_instruc)])
```

```python
print(chat_template1.messages[0].prompt.template)
```

```python
chat = ChatOpenAI(
    model = "gpt-3.5-turbo",
    temperature = 0,
    seed = 365,
    max_tokens = 30,
    openai_api_key = config.api_key
)
```

```python
list_output_parser = CommaSeparatedListOutputParser()
```

### And with that we have all the components necessary to build our first expression language chain a chat
### template, a chat model, and an output parser.

```python
chat_template_result = chat_template.invoke({'pet':'dog'})
```

```python
chat_result =chat.invoke(chat_template_result)
```

```python
list_output_parser.invoke(chat_result)
```

# Let's now apply the same steps as a one liner utilizing the Lang Chain expression language. Start by defining an object called chain first.
## | pipe symbol, which links elements in the expression language to show that the templates output will be the input for the subsequent component.


```python
chain =  chat_template | chat | list_output_parser 
```

```python
chain.invoke({'pet': 'rabbit'})
```

I hope you're starting to appreciate the elegance of this code.
It's simple, clean, and intuitive.
Moreover, it's easily traceable.
Should we hit an error, we can apply the invoke method to each component, study the output, and pinpoint
exactly where the code breaks.
