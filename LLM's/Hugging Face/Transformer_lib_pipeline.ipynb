## ZERO SHORT CLASSIFICATION 
### LETS SEE IF A LM, CAN CLASSIFIY SOME TEXT WITHOUT BEING EXPLICITY TRAINED ON THE CLASSISIFIACTIONS

```python
zeroshot_classification = pipeline("zero-shot-classification", model= "facebook/bart-large-mnli")
```

We need our sequence that we want to classify and we need our candidate labels.

These are the labels that we're going to classify our sentence into

```python
sequence_to_classify = "one day I will see the world"
candidate_labels = ['travel', 'cooking', 'dancing']
```

```python
zeroshot_classification(sequence_to_classify, candidate_labels)
```
