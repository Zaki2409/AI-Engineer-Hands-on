## Basically we are giving a model as param to the Autotokeizser() : The model name tells the tokenizer which "language" to speak.
 
### When you specify "bert-base-uncased", you're telling the tokenizer:

### "I want to talk to the BERT model, so please speak BERT's language!"

```python
from transformers import AutoTokenizer
```

```python
model= "bert-base-uncased"
```

```python
tokenizer = AutoTokenizer.from_pretrained(model)
```

```python
sentence = "Im so happy to be learing about LLM"
```

```python
# now to call the tokenizer on this sentecse, we have to craete input ids
input_ids = tokenizer(sentence)
print(input_ids)

#The token type IDs are used for some models, where we want to distinguish one sentence from another, or an input sequence and its relative classification.

#We can use numbers to distinguish one sentence from another, where tokens with the type ID of zero

#would belong to the first sentence, and token types with ID of one would be the second sentence or

#the classification.

#The attention mask essentially tells the model which words are important to pay attention to.
```

```python
# step 1
tokens = tokenizer.tokenize(sentence)
```

```python
print(tokens)
```

```python
token_ids = tokenizer.convert_tokens_to_ids(tokens)
```

```python
print(token_ids)
```

```python
decoded_ids = tokenizer.decode(token_ids)
print(decoded_ids)
```

## Using Hugging face and pytorch

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch
```

```python
print(sentence)
print(input_ids)
```

```python
tokenizer = AutoTokenizer.from_pretrained("distilbert-base-uncased-finetuned-sst-2-english")
input_ids_pt = tokenizer(sentence, return_tensors="pt")
print(input_ids_pt)
```

```python
#the output is the same, but it's now formatted within a PyTorch tensor.
```

### Lets now create our model for sequce classifaciton

```python
model = AutoModelForSequenceClassification.from_pretrained("distilbert-base-uncased-finetuned-sst-2-english")
```

```python
#We're now ready to use PyTorch on our inputs to do some inference of our own.
```

```python
#So with torch dot no grad we obtain the logits by running the model over our input IDs pt
with torch.no_grad():
    logits = model(**input_ids_pt).logits

# We then want to return the label of positive or negative for our input texts.

#So we'll get our predicted class ID and to use this we'll take argmax item.

# And then we take our model.config.id to label all this predicted class id.

predicted_class_id = logits.argmax().item()
model.config.id2label[predicted_class_id]  # This is essentially looking at the dictionary mapping of the numeric scores to the classification labels.

#So it's looking to see whether it should return positive or negative.
```
