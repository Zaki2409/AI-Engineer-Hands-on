# SIMPLE POETIC CHATBTO

```python
import openai
import config 
```

```python
api_key = config.api_key
```

```python
# to initialize openai client 
openai.api_key = api_key
```

```python
def poetic_chatbot(prompt):
    response = openai.ChatCompletion.create(
        model = "gpt-3.5-turbo",
        messages = 
[
            {
                "role": "system",
                "content": "You are a poetic chatbot."
            },
            {
                "role": "user",
                "content": "When was Google founded?"
            },
            {
                "role": "assistant",
                "content": "In the late '90s, a spark did ignite, Google emerged, a radiant light. By Larry and Sergey, in '98, it was born, a search engine new, on the web it was sworn."
            },
            {
                "role": "user",
                "content": "Which country has the youngest president?"
            },
            {
                "role": "assistant",
                "content": "Ah, the pursuit of youth in politics, a theme we explore. In Austria, Sebastian Kurz did implore, at the age of 31, his journey did begin, leading with vigor, in a world filled with din."
            },
            {
                "role": "user",
                "content": prompt
            }
        ],
        temperature = 1,
        max_tokens =20 )
    return response
```

```python
prompt = "When was chees made?"
poetic_chatbot(prompt)
```

```python
prompt = " What is the next course in the this udemy cousrse "
poetic_chatbot(prompt)

## the model gives the repsonse but not what the next course ?, this data insnt availabe by the model..
## GPT Models have been trained by a lot of data, but this data only goes as far as 2021,so they might not have the updated inforamtion ----->
### the solution LANGCHAIN , USING LANG CHAIN WE ARE ABLE TO IMPORT OUR OWN DATA AND HAVE IT READ BY OUR LLM, the the model can refrence this when creating a repoisnse usning tjhe langchian framework
```
