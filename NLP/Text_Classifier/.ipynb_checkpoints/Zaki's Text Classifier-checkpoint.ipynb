#### Our goal is to build an algo that can learn from this data and tehn classifiy new sentences by their sentiment


```python
import pandas as pd

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
```

```python
data = pd.DataFrame ([("i love spending time with my friends and family", "positive"),
("that was the best meal i've ever had in my life", "positive"),
("i feel so grateful for everything i have in my life", "positive"),
("i received a promotion at work and i couldn't be happier", "positive"),
("watching a beautiful sunset always fills me with joy", "positive"),
("my partner surprised me with a thoughtful gift and it made my day", "positive"),
("i am so proud of my daughter for graduating with honors", "positive"),
("listening to my favorite music always puts me in a good mood", "positive"),
("i love the feeling of accomplishment after completing a challenging task", "positive"),
("i am excited to go on vacation next week", "positive"),
("i feel so overwhelmed with work and responsibilities", "negative"),
("the traffic during my commute is always so frustrating", "negative"),
("i received a parking ticket and it ruined my day", "negative"),
("i got into an argument with my partner and we're not speaking", "negative"),
("i have a headache and i feel terrible", "negative"),
("i received a rejection letter for the job i really wanted", "negative"),
("my car broke down and it's going to be expensive to fix", "negative"),
("i'm feeling sad because i miss my friends who live far away", "negative"),
("i'm frustrated because i can't seem to make progress on my project", "negative"),
("i'm disappointed because my team lost the game", "negative")
                     ],
columns=['text', 'sentiment'])

# this data set will serve as the reaining ground for our logistic regression model
```

```python
#As you have seen earler, there are already many pre-trained sentiment analysis models avaliable, However sometimes
# their accuracy may not be high enough for you use case--- or you might have specific data that doesnt work well with 
#existing models---> in those caseses bulding your own custom model is a great solution
```

```python
# SHUFFLE THE DATA SET
#this meakes sure, we have a good mix of positive and negative sentences isntead of having them gourped
#LIKE ALL THE POSTIVE GORUPED FIRST AND THEN ALL THE NEGATIVE GROUEPD LATER
```

```python
data = data.sample(frac=1).reset_index(drop=True) # this mean facr =1 -> we are taking 100 % of the data but in a randomized order
```

```python
## PREPARING INPUTD FOR ALGO AND VECTORIZATION

x = data['text']
y = data['sentiment']

#CountVectorizer() -> Bag of words
# Each word = columns
# Each sentence = row
# Entries = word count
```

```python
countvec = CountVectorizer()
```

```python
countvec_fit = countvec.fit_transform(x) # here it klearns the voaclbualry and transfomes the sentencse into nuymber
```

```python
bag_of_words = pd.DataFrame(countvec_fit.toarray(), columns = countvec.get_feature_names_out())
# finally we will covert the bag of words matrix into pandas data frame.
# for the column names we use get feature names out, which returns the words from our vocabualry
# printing the dataframe shows us the bag of words represetation
```

```python
bag_of_words
```

```python
## SPLITING THE DATA INTO TRAINING AND TESTING SETS
```

```python
X_train, X_test, y_train, y_test = train_test_split(bag_of_words, y, test_size=0.3, random_state=7)
# we added the random_state = 7, so the split is reproducible --> we will alwasys get the same train test split each time 
# we run the code.

# radom_state() makes sure we have the same train and test split each time we run the code
```

```python
lr = LogisticRegression(random_state = 1).fit(X_train, y_train)
```

```python
y_pred_lr = lr.predict(X_test)
```

```python
accuracy_score(y_pred_lr, y_test)
```

```python
### not good performace.........however this is a good baseline model to use 
```

### classification_Report() : more detailed information on the performance

```python
print(classification_report(y_test ,y_pred_lr, zero_division=0))
```

# LETS IMPORVE THE ACCURACY, BY USING THE NAVE BAYES ALGO -> A SIMPLE CLASSIICATION ALGO THAT WORKS USING PROBABILITES

## NAIVIE BAYES COMBINE THE WORD COUNTS MATHEMATICALLY AND THEN PREDICTS THE MOST LIEKLLY CLASS FOR ANEW TEST

# its called naive becuz the algo assumes each word in a review is independent of the other

good for text anylas adn setimental snalysi

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
nb = MultinomialNB().fit(X_train, y_train)
```

```python
y_pred_nb = nb.predict(X_test)
```

```python
accuracy_score(y_pred_nb, y_test)
```

# lets again try to imporve using LINEA SUPOORT VECTOR MACHINE ALGO 
## the main idea behind this algo is to find the best possible boundry that separates the classes

use full for setimental anayaltiss and text classifacition

```python
from sklearn.linear_model import SGDClassifier
```

```python
svn = SGDClassifier().fit(X_train, y_train)
```

```python
y_pred_svn = svn.predict(X_test)
```

```python
accuracy_score(y_pred_svn, y_test)
```
