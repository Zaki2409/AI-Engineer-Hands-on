```python
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.corpus import stopwords
import re
import pandas as pd
```

```python
data = pd.read_csv("tripadvisor_hotel_reviews.csv")
```

```python
#Lower case
```

data['review_lowercase'] = data['Review'].str.lower()

```python
data.head()
```

```python
#Remove Stop Words
```

```python
en_stopwords = stopwords.words('english')
```

```python
en_stopwords.remove('not')
```

```python
#apply() : takes one column adn performs a custom opertaion on wevery calue in it COL BY COL, BUT WHEN aixs = 1...it means entire data row by row
data['review_no_stopwords'] = data['review_lowercase'].apply(lambda x:  ' '.join(word for word in x.split() if word not in (en_stopwords)))
```

```python
data['review_no_stopwords'][0]
```

```python
data['review_no_stopwords_no_punct'] = data.apply(lambda x: re.sub(r"[*]", "star", x['review_no_stopwords']),axis = 1)
```

```python
data['review_no_stopwords_no_punct'][2]
```

```python
data.head()
```

```python
#NOW remove any punctuations
data['review_no_stopwords_no_punct'] = data.apply(lambda x: re.sub(r"([^\w\s])", "", x['review_no_stopwords_no_punct']), axis=1)
```

```python
data.head()
```

```python
#Tokenization
data['tokenized'] = data.apply(lambda x: word_tokenize(x['review_no_stopwords_no_punct']), axis=1)
```

```python
data['tokenized'][0]
```

```python
#stemming
ps = PorterStemmer()
```

```python
data['stemmed'] = data['tokenized'].apply(lambda x: [ps.stem(token) for token in x])
```

```python
data.head()
```

```python
#Lemmatizer applied to compare with stemming
lemmatizer = WordNetLemmatizer()
```

```python
data['lemmatized']= data['tokenized'].apply(lambda y: [lemmatizer.lemmatize(token) for token in y]) 
```

```python
data['lemmatized'][0]
```

```python
data.head()
```

RIGHT NOW, EACH ROW IN THE 'LEMMATIZED' COKLUMN CONTAINS A SEPARATE LIST OF TOKENS--> EACH REVIEW IS STORES AS ITS OWN LIST OF LEMMATIZED WORDS. THEREFORE WE NEED TO COMBINE THESE SMALLER LISTS INTO ONE LONG LIST THAT CONTAINS EVERY TOKEN FROM ALL THE REVIEWS

```python
token_clean = sum(data['lemmatized'], [] )  
#In python,when we use sum() with lists instead of numbers, it joins them together
```

```python
#most frequntyly occuring unigrams
unigrams = (pd.Series(nltk.ngrams(token_clean, 1)).value_counts())
print(unigrams)
```

```python
#most frequntyly occuring bigrams
bigrams = (pd.Series(nltk.ngrams(token_clean, 2)).value_counts())
print(bigrams)
```

these gave a nice idea of topics and themes that are comming up in the data, without requiring difficult or time analysis
